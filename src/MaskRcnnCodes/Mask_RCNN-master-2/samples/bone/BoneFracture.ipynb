```
from samples.bone import bone
import mrcnn.model as modellib
# Configurations
config = bone.BoneConfig()
config.display()
# Create model
model = modellib.MaskRCNN(mode="training", config=config,
                                model_dir="log")

# Load weights
model.load_weights("../mask_rcnn_coco.h5", by_name=True, exclude=[
        "mrcnn_class_logits", "mrcnn_bbox_fc",
        "mrcnn_bbox", "mrcnn_mask"])
bone.train(model)
```

```
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from mrcnn.visualize import display_instances
from mrcnn.config import Config
from mrcnn.model import MaskRCNN

# define 81 classes that the coco model knowns about
class_names = ['BG', 'fracture']

# define the test configuration
class TestConfig(Config):
     NAME = "test"
     GPU_COUNT = 1
     IMAGES_PER_GPU = 1
     NUM_CLASSES = 1 + 1

# define the model
rcnn = MaskRCNN(mode='inference', model_dir='./', config=TestConfig())
# load coco model weights
rcnn.load_weights('drive/My Drive/mask_rcnn_bone_0030.h5', by_name=True)
# load photograph
img = load_img('drive/My Drive/0021.png')
img = img_to_array(img)
# make prediction
results = rcnn.detect([img], verbose=0)
# get dictionary for first prediction
r = results[0]



# show photo with bounding boxes, masks, class labels and scores
display_instances(img, r['rois'], r['masks'], r['class_ids'], class_names, r['scores'])
```
